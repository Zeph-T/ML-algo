```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("data.csv")
```

```python
X=data["x"]
y=data["y"]
plt.scatter(X,y,color="red")
```

```python
msk=np.random.rand(len(data))<0.8
```

```python
train=data[msk]
test=data[~msk]
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
xx=np.asanyarray(train[["x"]])
yy=np.asanyarray(train[["y"]])
model.fit(xx,yy)
model.intercept_
```

```python
plt.scatter(train.x,train.y,color="blue")
plt.plot(xx,model.coef_*xx+model.intercept_,"-r")
```

```python
from sklearn.metrics import r2_score
test_x=np.asanyarray(test[["x"]])
test_y=np.asanyarray(test[["y"]])
test_y_hat=model.predict(test_x)
print(r2_score(test_y_hat,test_y))
```
